# 4 Translated Function

```python
def translate_text(text, source_lang, target_lang):
    if not text.strip():
        return "Please enter some text."

    try:
        return GoogleTranslator(
            source=Languages[source_lang],
            target=Languages[target_lang]
        ).translate(text)

    except Exception as e:
        return f"Error: {e}"
```

# 5 Swap Language Function

```python
def swap(source_lang, target_lang, input_text, output_text):
    if source_lang == "Auto Detect":
        return (
            source_lang,
            target_lang,
            output_text,
            input_text
        )
    
    return (
        target_lang, 
        source_lang, 
        output_text,
        input_text
    )
```

# 6 Clear Function

```python
def clear_text():
    return "", ""
```

# 7 Build Gradio UI

```python
css = """
footer{
    display: none !important;
}

.gradio-container{
    max-width: 1100px;
    margin: auto;
}
"""

with gr.Blocks(
    theme = gr.themes.Soft(),
    css=css
) as demo:
    gr.Markdown("""
    <div style="text-align: center; font-family: Arial, sans-serif; font-weight: bold;">
        <h1>🌐 AI-Powered Language Translation Tool</h1>
        <p>Translate text across 50+ languages instantly</p>
    </div>
    """)
    
    input_text = gr.Textbox(
        placeholder="Enter text to translate",
        label="Input Text",
        lines=5,
        show_copy_button=True
    )
    
    with gr.Row():
        source_lang = gr.Dropdown(
            choices=list(Languages.keys()),
            value="Auto Detect",
            label="Source Language"
        )
        
        target_lang = gr.Dropdown(
            choices=list(Languages.keys())[1:],
            value="English",
            label="Target Language"
        )
        
    with gr.Row():
        swap_btn = gr.Button("🔄 Swap Language")
        translate_btn = gr.Button("🌍 Translate")
        clear_btn = gr.Button("🧹 Clear Text")

    output_text = gr.Textbox(
        placeholder="Translated text will appear here",
        label="Translated Text",
        lines=5,
        show_copy_button=True
    )
    
    gr.Examples(
        examples=[
            ["Hello, how are you?"],
            ["Good morning!"],
            ["I love artificial intelligence."],
            ["Welcome to CodeAlpha."]
        ],
        inputs=input_text
    )
    
    translate_btn.click(
        fn=translate_text,
        inputs=[input_text, source_lang, target_lang],
        outputs=output_text,
    )
    
    swap_btn.click(
        fn=swap,
        inputs=[
            source_lang,
            target_lang,
            input_text,
            output_text
        ],
        outputs=[
            source_lang,
            target_lang,
            input_text,
            output_text
        ]
    )
    
    clear_btn.click(
        fn=clear_text,
        inputs=[],
        outputs=[input_text, output_text]
    )
    
    gr.Markdown("""
    <div style="text-align: center; font-family: Arial, sans-serif; font-weight: bold;">
        <p>❤️ Developed by Ragavan <br>
        Built with Python, Gradio, and deep-translator</p>
    </div>
    """)
    
    
```

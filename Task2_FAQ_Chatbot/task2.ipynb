# Step 4 Preprocess and Vectorize Text

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
df["preprocessed_question"] = df["question"].apply(preprocess_text)
df[["question", "preprocessed_question"]].head()
```

```python
vectorizer = TfidfVectorizer(
    stop_words='english',
    lowercase=True,
    ngram_range=(1, 2)
)

tfidf_matrix = vectorizer.fit_transform(df["preprocessed_question"])

print(f"TF-IDF matrix shape: {tfidf_matrix.shape}")
```

```python
print(df.head())
print("Vocabulary Size:", len(vectorizer.get_feature_names_out()))
```

# Step 5 Preprocess User Query

```python
def preprocess_query(text):
    processed = preprocess_text(text)
    query_vector = vectorizer.transform([processed])
    return processed, query_vector

processed_query, query_vector = preprocess_query("What are the symptoms of depression?")

print("Processed Query:", processed_query)
print("Query Vector Shape:", query_vector.shape)
```

# Step 6 Compute Cosine Similarity

```python
similarities = cosine_similarity(query_vector, tfidf_matrix)

best_match_idx = similarities.argmax()

best_score = similarities[0, best_match_idx]

thereshold = 0.30

if best_score >= thereshold:
    print("Best Matching Question:")
    print(df.loc[best_match_idx, "question"])
else:
    print("No relevant answer found.")
```

```python
print("Best Matching Question:", best_match_idx)
print(df.loc[best_match_idx, "question"])

print("\nBest Answer:")
print(df.loc[best_match_idx, "answer"])

print("\nSimilarity Score:", best_score)
```

# Step 7 Generate Best Matching Response

```python
def get_response(user_query, thereshold=0.30):
    processed_query = preprocess_text(user_query)
    
    query_vector = vectorizer.transform([processed_query])
    
    similarities = cosine_similarity(query_vector, tfidf_matrix)
    
    best_match_idx = similarities.argmax()
    best_score = similarities[0, best_match_idx]
    
    if best_score >= thereshold:
        return {
            "question": df.loc[best_match_idx, "question"],
            "answer": df.loc[best_match_idx, "answer"],
            "score": best_score
        }
    else:
        return {
            "question": None,
            "answer": "I'm sorry, I couldn't find a relevant answer to your question.",
            "score": best_score
        }
```

```python
result = get_response("What is stomach disorder?")

print("Matched Question:", result["question"])
print("Answer:", result["answer"])
print("Similarity Score:", result["score"])
```

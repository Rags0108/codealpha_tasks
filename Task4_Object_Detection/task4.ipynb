# Step 2 Import Libraries

```python
import cv2
import numpy as np
from ultralytics import YOLO
```

# Step 3 Load YOLO model

```python
model = YOLO('models/yolo11n.pt')

model.info()

print(model.names)
print("Model loaded successfully!")
```

# Step 4 Open webcam or video

```python
use_cam = True

video_path = "videos/video.mp4"

if use_cam:
    cap = cv2.VideoCapture(3) # this index 3 it's my system webcam index 
    print("Using Webcam")
else:
    cap = cv2.VideoCapture(video_path)
    print(f"Using Video File:{video_path}")
```

```python
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Some webcams return 0 FPS, so use a default
if fps == 0:
    fps = 30

# Create video writer
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(
    "videos/output.mp4",
    fourcc,
    fps,
    (width, height)
)
```

```python
print(f"Video Source Ready")
print(f"Resolution: {width}x{height}")
print(f"FPS: {fps:2.0f}")
```

# Step 5: Run object detection

```python
while cap.isOpened():

    success, frame = cap.read()

    if not success:
        print("Video processing completed.")
        break

    results = model.track(
        frame,
        persist=True,
        tracker="bytetrack.yaml",
        conf=0.5,
        iou=0.5,
        verbose=False
    )

    counts = {}

    for result in results:
        for cls_id in result.boxes.cls:
            cls_name = model.names[int(cls_id)]
            counts[cls_name] = counts.get(cls_name, 0) + 1

    total_count = sum(counts.values())
    counts["Total"] = total_count

    annotated_frame = results[0].plot()
    
    y = 30
    for cls_name, count in counts.items():

        cv2.putText(
            annotated_frame,
            f"{cls_name}: {count}",
            (10, y),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0),
            2
        )
        y += 30

    out.write(annotated_frame)

    cv2.imshow("Object Detection & Tracking", annotated_frame)

    print("Current Object Counts:", counts)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Stopped by user.")
        break

cap.release()
cv2.destroyAllWindows()

print("Final Object Counts:", counts)
```
